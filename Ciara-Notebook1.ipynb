This notebook analyzes [WPRDC - Pittsburgh Police Arrest Data](https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f) which is saved and refered to as "Arrests.csv" in this repository. 

The data set from WPRDC has arrest data starting 8/26/2016 to 8/27/2023 and many different data columns including personal identifying information, incident type, location, time, data, and more. This notebook is focused on the Incident Neighboorhood data column and so the code only filters through this column. Additionally, in order to keep the data metrics consisent across the entire repository, this notebook exclusivey looks at arrests from 2020, which is filtered out within the code.

This notebook looks to determine which neighboorhood is the best in Pittsburgh based on how many arrests have occured within each neighborhood. 
The least amount of arrests will be considered the best in this notebook. 

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#read in the file
df = pd.read_csv("Arrests.csv")

#seperate the data so it is all 2020 arrests
querymask = df['ARRESTTIME'].str.contains('2020')
arrests = df[querymask]

groups = arrests.groupby(['INCIDENTNEIGHBORHOOD']).size()
#plot the data in a bar chart 
groups.plot.bar(figsize=(150,50))

#change the y and x ticks
plt.xticks(rotation = 45, ha = 'right', rotation_mode = 'anchor', fontsize=50)
plt.yticks(fontsize = 50)
plt.style.use('seaborn-dark')

#add a title and labels to the graph
plt.title("Pittsburgh Arrests in 2020", fontsize=150)
plt.xlabel("Neighborhoods", fontsize=80)
plt.ylabel("Arrests", fontsize=80)
```

#### Initial Reaction

The data filtering done in the code allows all of the neighborhoods to be represented by the number of arrests made for side-by-side comparison in a bar chart. The bar chart clearly shows thr divide between the highest and lowest, but does not have a clear "winner" of least arrests so to speak. The bar chart shows Chartiers City, Mt. Oliver Boro, New Homestead, Outside County, Regent Square, and St. Clair as having the lowest amoutn of arrests, it is not easy to tell which has the highest population. 

So to continue this analysis, the exact arrest numbers will be extracted from the chart to determine which neighborhood is the best.

```python
#print the specific numbers of the arrests DF to compare
print('Chartiers City: ', groups.get('Chartiers City', 0))
print('Mt. Oliver Boro: ', groups.get('Mt. Oliver Boro', 0))
print('New Homestead: ', groups.get('New Homestead', 0))
print("Outside County: ", groups.get('Outside County', 0))
print("Regent Square: ", groups.get('Regent Square', 0))
print("St. Clair: ", groups.get('St. Clair', 0))
```

There are 4 neighborhoods that have the exact same amount of arrests. It is a very low number for all years, which means these for neighborhoods are very safe.
