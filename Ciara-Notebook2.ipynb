This notebook analyzes [WPRDC - City of Pittsburgh Neighborhood Population 2020](https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f) which is saved and refered to as "Neighborhood_Population.csv" in this repository. 

The data set from WPRDC has population data from each demographic in 2010 and 2020. In order to stay consistent with the other notebooks in this repository, only the 2020 population data will be considered in this notebook. Additionally, as this notebook is aiming to look at total population, the seperated demoraphic populations listed in this data set are not considered in this data set, only the total poulation column.

This notebook looks to determine which neighboorhood is the best in Pittsburgh based on the total population in each neighborhood. 
The neighborhood with the highest population will be the best, as the higher population is an indicator that more people want to live there. 

```python
#import panda and others to let the code run propperly
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#save the data from the file 
df = pd.read_csv("Neighborhood_Population.csv")

#save the names of all the neighborhoods
neighborhood_names = []
for index, row in df.iterrows():
    name = row['Neighborhood']
    neighborhood_names.append(name)

#save the population data 
neighborhood_pop = []
for index, row in df.iterrows():
    pop = row['2020_Total_Population']
    neighborhood_pop.append(pop)
    
#create a new Series with just the important information
population_df = pd.Series(neighborhood_pop, index = neighborhood_names)

#plot the data in a bar chart 
population_df.plot.bar(figsize=(150,50))
    
#change the xticks and yticks
plt.xticks(rotation = 45, ha = 'right', rotation_mode = 'anchor', fontsize=50)
plt.yticks(fontsize = 50)
plt.style.use('seaborn-dark')

#add a title and labels to the graph
plt.title("Pittsburgh Neighborhood Populations in 2020", fontsize=150)
plt.xlabel("Neighborhoods", fontsize=80)
plt.ylabel("Population Size", fontsize=80)
```

#### Initial Reaction

The data filtering done in the code allows all of the neighborhoods to be represented by their population for side-by-side comparison in a bar chart. The bar chart allows the visuals of which neighborhoods have the highest and lowest, and then allows for further filtering to be done because of this display. The bar chart shows Shadyside and Squirrel Hill South as the highest populations, but because the bars are so close, it is not easy to tell which has the highest population. 

So to continue this analysis, the exact population numbers will be extracted from the chart to determine which neighborhood is the best.

```python
#count variable for the location 
count = 0
#filter through to find the location of the top neighborhoods
for index, row in df.iterrows():
    name = row['Neighborhood']
    if (name == "Shadyside"):
        SSloc = count
    if (name == "Squirrel Hill South"):
        SHSloc = count
    count = count +1
#use the locations to print the specific rows of the Series to compare
print('Shadyside: ', population_df.iloc[SSloc])
print("Squirrel Hill South: ", population_df.iloc[SHSloc])
```

The population numbers for the neighborhoods that seemed to be the highest, Squirrel Hill South and Shadyside, are the exact same! Therefore, because no other data than the 2020 Total Population is being used in this notebook, there are two best neighborhoods.
