```python
#Importing pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Loading data
assessments = pd.read_csv("Property-Assessments.csv", index_col = "_id")
df = pd.read_csv("Property-Assessments.csv")
```

```python
groups = df[['SCHOOL_DISTRICT', 'CURRENT TOTAL VALUE']].copy()
groups.plot.bar(figsize =(40,30))
```

```python
assessments.head(5)
```

```python
#Plot Data
df["CURRENT TOTAL VALUE"].value_counts().sort_index().plot(kind='bar', rot=90, figsize = (10,5))
```

```python
# filter to important info
year_mask = df['TAX YEAR'] == 2020
year = df[year_mask]
year.plot()

print(year)

value = year.groupby("SCHOOL_DISTRICT").sum()['CURRENT TOTAL VALUE']
value.sort_values(ascending=False)
```

```python
year_mask = df['TAX YEAR'] == 2020
year = df[year_mask]
year.head()
```

```python
year_mask = df['TAX YEAR'] == 2020
year = df[year_mask]
value = year.groupby("MUNI_NAME").sum()['CURRENT TOTAL VALUE']
value.sort_values(ascending=False)
```
