This note book analyzes survey data from different aged homeowners from different Pittsburgh neighborhoods. The data is refered to as "Housing.csv" in the repository. 

Source - https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2014-miscellaneous-data/resource/262d10d1-1e49-41f6-bb8a-0c4511a9085c

This data set is from WPRDC and was obtained from the Housing Question 19 in the 2013 America Community Survey (ACS). The question allowed for data on the estimate value's of homes (which could be either land property, a house, apartment, mobile home, etc.) if they were to sold and allowed home owners to be groupdc of adults under 25, 25-44, 45-65, and 65 and older. As whole this data allows poeple to determine the quality of homes in different neighborhoods, if housing in Pittsburgh is affordable, and more. In this notebook, I will specifically be looking at the number of homeowners per area. I believe that if there are more people living in more area, it allows for a stronger community and more social events.

Thus the neighborhood with the highest number of homeowners would be the best neighborhood in Pittsburgh to live. 

```python
#link to data https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2014-miscellaneous-data/resource/262d10d1-1e49-41f6-bb8a-0c4511a9085c

#Importing pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#Reading data & assigning a value
data = pd.read_csv("Housing.csv",index_col=1)
prices = pd.Series(data["Estimate; Total:"])

#Creating Visual
prices.plot.bar(figsize=(30,20))

#Adjusting Sizing, Adding Labels, & other customization
plt.xticks(rotation = 45, ha = 'right', rotation_mode = 'anchor')
plt.yticks(fontsize = 20)
plt.style.use('seaborn-dark')

plt.title("Estimate Total of Home Owners by Neighborhood", fontsize = 50)
plt.xlabel("Pittsburgh Neighborhoods", fontsize = 25)
plt.ylabel("Average Number of Home Owners", fontsize = 25)

#Best neighborhood is determined by which neighborhood has the most homeowners
#From this data it can be determined that Shadyside is the best neighborhood because it has the most homeowners. 
```

#### Initial Reaction

From filtering the data to visualize the estimate number of homeowners per neighborhood into a bar chart it can easily be seen with neighborhoods have many homeowners and which have few. As an overall view most neighborhoods seem to have the same amount of homeowners, but certain neighborhoods stick out for having more or less than other. The neighborhoods with the greatest number of homeowners include Shadyside, Squirrel Hill South, and Brookline while neighborhoods like Chateau, South Shore, St. Clair, and West End. 

```python
#Highlight the neighborhoods with the most homeowners
print("Ranking:")
print('1. Shadyside: ', prices.get('Shadyside', 0))
print('2. Squirrel Hill South: ', prices.get('Squirrel Hill South', 0))
print('3. Brookline: ', prices.get('Brookline', 0))
```

#### The Best Neighborhood

From this data visualization Shadyside is the best neighborhood in Pittsburgh because it has the most homeowners.
