Perspective based on property assessments in the pittsburgh neighborhoods. 

Source - https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2014-miscellaneous-data/resource/262d10d1-1e49-41f6-bb8a-0c4511a9085c

```python
#link to data https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2014-miscellaneous-data/resource/262d10d1-1e49-41f6-bb8a-0c4511a9085c

#Importing pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#Reading data & assigning a value
data = pd.read_csv("Housing.csv",index_col=1)
prices = pd.Series(data["Estimate; Total:"])

#Creating Visual
prices.plot.bar(figsize=(30,20))

#Adjusting Sizing, Adding Labels, & other customization
plt.xticks(rotation = 45, ha = 'right', rotation_mode = 'anchor')
plt.style.use('seaborn-dark')

plt.title("Estimate Total of Home Owners by Neighborhood", fontsize = 40)
plt.xlabel("Pittsburgh Neighborhoods", fontsize = 30)
plt.ylabel("Average Number of Home Owners", fontsize = 30)

#Best neighborhood is determined by which neighborhood has the most homeowners
#From this data it can be determined that Shadyside is the best neighborhood because it has the most homeowners. 
```
